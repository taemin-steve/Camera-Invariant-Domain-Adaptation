```python
import cv2
import numpy as np
import os
import glob
```

```python
train_target = glob.glob("../Data/train_target_image/*")
train_target.sort()
train_target
```

```python
img = cv2.imread(train_target[0])
```

```python
accumulator = np.zeros_like(img, dtype=np.float64)
alpha = 1/len(train_target)

for img_path in train_target:
    current_img = cv2.imread(img_path)
    accumulator += (alpha * current_img)

cv2.imwrite("combined_image.png", accumulator)
    
```

```python
image = cv2.imread("./combined_image.png", cv2.IMREAD_GRAYSCALE)

# Apply Canny edge detection
low_threshold = 10  # Adjust as needed
high_threshold = 150  # Adjust as needed
edges = cv2.Canny(image, low_threshold, high_threshold)

# Save the edge-detected image
cv2.imwrite("edge_detected_image.png", edges)
```
